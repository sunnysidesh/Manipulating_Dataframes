```
import pandas as pd
census_population = pd.read_csv("state-populations.csv")
census_regions = pd.read_csv("census-divisions.csv")

print(census_population.columns)
print(census_regions.columns)
```

```
census_population.reset_index(level = 0, inplace = True)
```

```
census_population.melt(id_vars = ['2010', '2011', '2012', '2013', '2014', '2015', '2016'], var_name = 'year', value_name = 'population')
```

```
census_population.melt(id_vars = 'state', var_name = 'year', value_name = 'population')
```

```
census = pd.merge(census_population, census_regions)

print(census.columns)
print(census.shape)
```

```
census_no_duplicates = census.drop_duplicates(['state'])
census_no_duplicates
```

```
census_no_duplicates.groupby(['2010', '2011', '2012', '2013', '2014', '2015', '2016']).describe()
```
